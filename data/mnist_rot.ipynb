# MNIST-ROT Dataset

```python
import numpy as np
import loader
import visualization
```

### Loading the Data

```python
train, valid, test = loader.mnist_rot_np()

train_x, train_y = train['x'], train['y']
valid_x, valid_y = valid['x'], valid['y']
test_x, test_y = test['x'], test['y']
```

### Dataset Size

```python
print('training samples:', f'{len(train_x):,}')
print('validation samples:', f'{len(valid_x):,}')
print('test samples:', f'{len(test_x):,}')
```

### Data Inspection

```python
print('input shape:', train_x[0].shape, '-> flattened 28x28 image')
print('output_shape:', train_y[0].shape, '-> directly the corresponding digit')
```

```python
print(f'input value range: {np.min(train_x).round(2)} - {np.max(train_x).round(2)}')
```

```python
print('number of classes:', len(np.unique(train_y)))
print('classes:', list(np.unique(train_y)))
```

```python
print('digit count:')
print(*[f'{y}: {n}' for y, n in zip(*np.unique(train_y, return_counts=True))], sep='\n')
```
